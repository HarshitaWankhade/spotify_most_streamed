```python
data2_col=['artist_count',
 'released_year',
 'released_month',
 'released_day',
 'in_spotify_charts',
 'in_apple_playlists',
 'in_apple_charts',
 'in_deezer_charts',
 'bpm',
 'danceability_%',
 'valence_%',
 'energy_%',
 'acousticness_%',
 'instrumentalness_%',
 'liveness_%',
 'speechiness_%',
 'streams',
 'in_deezer_playlists',
 'in_shazam_charts',
 'key',
 'mode','in_spotify_playlists']
```

```python
data2 = data2.reindex(columns=data2_col)
```

```python
data2.head()
```

```python
X=data2.iloc[:,:-1]
X.head(9)
```

```python
list(X)
```

```python
y=data2.iloc[:,-1:]
y.head(9)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=27,random_state=42)
from sklearn.tree import DecisionTreeRegressor

model=DecisionTreeRegressor(random_state=42)
model.fit(X,y)
```

```python
from sklearn.metrics import accuracy_score

ac_test=model.score(X_test,y_test)
ac_train=model.score(X_train,y_train)

print(ac_test,ac_train)
```
